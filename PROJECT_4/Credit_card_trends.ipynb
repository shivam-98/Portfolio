## Project :- Automating Customer Details

### Problem Statement :- 
The bank is interested to automate the process of discovering the trends in the customer credit card.

### Dataset Description :-
The sample Dataset summarizes the usage behavior of 8950 active credit card holders and have 18 behavioral features.

File Format: CSV

Rows: 8950,
Columns: 18

### Feature Description :-

<b>CUST_ID</b> : Identification of Credit Card holder (Categorical)
    
<b>BALANCE</b> : Balance amount left in their account to make purchases 
    
<b>BALANCE_FREQUENCY</b> : How frequently the Balance is updated, score between 0 and 1 (1 = frequently updated, 0 = not frequently updated) 
    
<b>PURCHASES</b> : Amount of purchases made from account
    
<b>ONEOFF_PURCHASES</b> : Maximum purchase amount done in one-go 
    
<b>INSTALLMENTS_PURCHASES</b> : Amount of purchase done in installment 
    
<b>CASH_ADVANCE</b> : Cash in advance given by the user 
    
<b>PURCHASES_FREQUENCY</b> : How frequently the Purchases are being made, score between 0 and 1 (1 = frequently purchased, 0 = not frequently purchased)
    
<b>ONEOFFPURCHASESFREQUENCY</b> : How frequently Purchases are happening in one-go (1 = frequently purchased, 0 = not frequently purchased) 
    
<b>PURCHASESINSTALLMENTSFREQUENCY</b> : How frequently purchases in installments are being done (1 = frequently done, 0 = not frequently done) 
    
<b>CASHADVANCEFREQUENCY</b> : How frequently the cash in advance being paid 
    
<b>CASHADVANCETRX</b> : Number of Transactions made with "Cash in Advanced" 
    
<b>PURCHASES_TRX</b> : Numbe of purchase transactions made 
    
<b>CREDIT_LIMIT</b> : Limit of Credit Card for user 
    
<b>PAYMENTS</b> : Amount of Payment done by user 
    
<b>MINIMUM_PAYMENTS</b> : Minimum amount of payments made by user 
    

<b>PRCFULLPAYMENT</b> : Percent of full payment paid by user 
    
<b>TENURE</b> : Tenure of credit card service for user

### Removing Outliers

Making a list of columns whose outliers have to be removed

```python
cnames = ['CREDIT_LIMIT', 'PAYMENTS', 'MINIMUM_PAYMENTS']
cnames
```

```python
def removeOutliers(data, cnames):
    for i in cnames:
        print(i)
        q75, q25 = np.percentile(data.loc[:,i], [75 ,25])
        iqr = q75 - q25

        min=q25 - (iqr*1.5)
        max=q75 + (iqr*1.5)
    
        data = data.drop(data[data.loc[:,i]<min].index)
        data = data.drop(data[data.loc[:,i]>max].index)
    return data
```

```python
df1 = removeOutliers(df1, cnames)
```

### Choosing Optimal Clusters for k

```python
from sklearn.cluster import KMeans
k=list(range(1,10))
wcss=[]# wcss=within cluster sum of squred error
for i in k:
    km=KMeans(n_clusters=i)
    km.fit(df3)
    wcss.append(km.inertia_)
plt.xlabel(" k value")
plt.ylabel(" Within cluster sum of suqred error(wcss)")
plt.title(" Finding best k value using elbow method")
plt.plot(k,wcss,marker='o')

```

##### The optimal clusters for k can be either 3,4,5 because after 5 the plot looks to become linear

### In order to select among the above three, we use Silhouette method

#### This method measures how similar a point is to it's own cluster compared to other clusters. Greater the value, good at measuring the cluster

```python
from sklearn import metrics
for i in range(3,6):
    kmeans_labels=KMeans(n_clusters=i,random_state=123).fit_predict(df3)
    print("Silhouette score for {} clusters k-means : {} ".format(i,metrics.silhouette_score(df3,kmeans_labels, metric='euclidean').round(3)))
```

##### For k=5 we are getting high Silhouette score

```python
kmeans = KMeans(5)
kmeans.fit(df3)
labels = kmeans.labels_
labels
np.unique(labels)
```

```python
clusters=pd.concat([df3, pd.DataFrame({'cluster':labels})], axis=1)
clusters.head(10)
```

### The labels of the clusters are added to the original data in order to identify for which cluster they belong.

```python
df2['k-means cluster']=labels
df2.groupby('k-means cluster').mean().T
```

## K-Means using PCA (no correlation)

```python
from sklearn.decomposition import PCA
pc = PCA(n_components=10).fit(data_standardized) #selecting components variance as above 90% and the optimal components are 10
p = pd.Series(pc.explained_variance_ratio_.cumsum()).plot()
p.set_xlabel('No. of principal components')
p.set_ylabel('Cummulative Explained Variance Ratio')
pc = PCA(n_components=10).fit_transform(data_standardized) #fitting and transforming the standardized data with 10 components
```

### Selecting optimal clusters for K

```python
from sklearn.cluster import KMeans
k=list(range(1,10))
wcss=[]# wcss=within cluster sum of squred error
for i in k:
    km=KMeans(n_clusters=i)
    km.fit(pc)
    wcss.append(km.inertia_)
plt.xlabel(" k value")
plt.ylabel(" Within cluster sum of suqred error(wcss)")
plt.title(" Finding best k value using elbow method")
plt.plot(k,wcss,marker='o')
```

##### Here also we need to choose the clusters among 3,4,5 because after 5 the plot tends to become linear

### Using Silhouette score to pick the number of clusters among the above three

```python
from sklearn import metrics
for i in range(2,6):
    kmeans_labels=KMeans(n_clusters=i,random_state=123).fit_predict(pc)
    print("Silhouette score for {} clusters k-means : {} ".format(i,metrics.silhouette_score(pc,kmeans_labels, metric='euclidean').round(3)))
```

##### For k=3 we are getting high score. So we will choose number of clusters as 3.

## Applying DB-Scan Clustering

```python
from sklearn.cluster import DBSCAN
results = pd.DataFrame(columns = ['Eps','Min_Samples', 'Number of Cluster', 'Silhouette Score'])
```

#### There are two important parameters in DB-Scan algorithm. Those are ,
#### 1. Epsilon
#### 2. Minimum number of samples

##### Epsilon :- This parameter determines the neighborhood around the point. i.e., if the distance between any two points is less than or equal to epsilon then they are considered as part of one cluster.

##### Min_Samples :- This parameter determines the minimum number of data points required within the epsilon value (radius).

### For correlation features

```python

from sklearn.cluster import DBSCAN
results = pd.DataFrame(columns = ['Eps','Min_Samples', 'Number of Cluster', 'Silhouette Score'])
for i in range(1,5):
    for j in range(1,5):
        dbscan_cluster = DBSCAN(eps = i, min_samples = j)
        clusters = dbscan_cluster.fit_predict(df3)
        if len(np.unique(clusters))>2:
            results = results.append({'Eps':i*0.2,
                                     'Min_Samples':j,
                                     'Number of Cluster':len(np.unique(clusters)),
                                     'Silhouette Score':metrics.silhouette_score(df3,clusters)}, ignore_index = True)

```
